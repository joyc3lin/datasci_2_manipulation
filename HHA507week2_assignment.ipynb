```python
import pandas as pd
import re
import numpy as np
```

# **Loading the Dataset**

```python
df = pd.read_csv('https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv')
df
```

## **1. Data Cleaning with Pandas**

## Identifying and handling the missing values

```python
#converting missing to null
df.replace('missing', pd.np.nan , inplace=True)
df
```

```python
#checking cell types
print(df['Patient Age'].isnull().sum())

for c in df.columns:
  print(f"{c}: {type(df[c][0])}")
```

```python
#code for replacing missing cell values with average of column
#df['Patient Age'] = pd.to_numeric(df['Patient Age'], errors = "coerce")
#df['Patient Age'].fillna(df['Patient Age'].mean(), inplace=True)
#df
```

```python
#dropping rows with missing data
df.dropna(inplace=True)
print(df.shape)
df
```

## Removing duplicate rows and columns

```python
df.drop_duplicates(inplace=True)
print(df.shape)
```

```python
df.loc[:,~df.T.duplicated(keep='first')]
print(df.shape)
```

## Cleaning Column Names


```python
# Function to remove white space and special characters from a value
def clean_value(value):
    cleaned_value = re.sub(r'\s+|[^a-zA-Z0-9]', '', str(value))
    return cleaned_value

# Apply the clean_value function to all columns
#for column in df.columns:
    #df[column] = df[column].apply(clean_value)

# renames headers
df.rename(mapper=clean_value, axis=1, inplace=True)

df
```

# **2. Data Transformations**

## creating new column

```python
# turning values in PaymentDue, PrescriptionCost, and AverageBP into floats
df['PaymentDue'] = pd.to_numeric(df['PaymentDue'], errors = "coerce")
df['PrescriptionCost'] = pd.to_numeric(df['PrescriptionCost'], errors = "coerce")
df['AverageBP'] = pd.to_numeric(df['AverageBP'], errors = "coerce")
```

```python
# adding TotalPaymentDue column that adds values of PaymentDue and PrescriptionCost columns
df['TotalPayment'] = df['PaymentDue'] + df['PrescriptionCost']
df
```

```python
# adding column BpStatus that shows if patient BP is  normal, or abnormal
df['BPStatus'] = np.where((df['AverageBP'] > 90) & (df['AverageBP'] < 120), 'Normal', 'Abnormal')
df
```

## Aggregate data using groupby and compute summary statistics

```python
# summary statistic to show the preferred doctors of the pateitns
df.groupby(['PreferredDoctor']).describe()
```

```python
#table showing averageBP based on patient gender
grouped_df = df.groupby('Gender').agg({
    'AverageBP': 'mean'
})
grouped_df
```

## Use pivot tables or cross-tabulations for multi-dimensional analysis

```python
# table showing total cost of prescriptions for medications prescribed to disease disgnosed
pivot_df = df.pivot_table( index='DiseaseDiagnosed', columns='MedicationPrescribed', values = 'PrescriptionCost', aggfunc='sum' )
pivot_df
```

```python
dataset_path = 'https://raw.githubusercontent.com/hantswilliams/HHA_507_2023/main/WK2/data/healthcare_data_cleaning.csv'
```

```python
# Using Pandas
start_time = time.time()
pandas_df = pd.read_csv(dataset_path)
pandas_time = time.time() - start_time

#Using polars
start_time2 = time.time()
polars_df = pl.read_csv(dataset_path)
polars_time = time.time() - start_time2

# Print results and execution times
print(f'The time it took for pandas to load: {pandas_time:.4f} seconds')
print(f'The time it took for polars to load: {polars_time:.4f} seconds')
```

**Observations**: datasets loaded in with polars take much less time than datasets loaded in with pandas
